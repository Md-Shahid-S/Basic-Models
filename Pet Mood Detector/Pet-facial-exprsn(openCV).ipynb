```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
```

```python
model = load_model('H:\My Drive\Basic Models\Models\pet_mood_model.h5')
```

```python
class_labels = ['Angry', 'Happy', 'Other', 'Sad']  
```

```python
cap = cv2.VideoCapture(0)  # 0 = default camera

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Resize and preprocess the frame
    img = cv2.resize(frame, (128, 128))
    img = img.astype("float32") / 255.0
    img = np.expand_dims(img, axis=0)  # Model expects 4D input: (1, 128, 128, 3)

    # Prediction
    pred = model.predict(img)
    mood = class_labels[np.argmax(pred)]

    # Display the result
    cv2.putText(frame, f"Pet Mood: {mood}", (20, 40),
                cv2.FONT_HERSHEY_SIMPLEX, 1.1, (0, 255, 0), 2)
    
    cv2.imshow("Pet Mood Detector", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):  # Quit on pressing 'q'
        break

cap.release()
cv2.destroyAllWindows()
```
